# Analyzing Netflix titles

- This is assignment 1. In this notebook, we will be answering how has the number of titles added to Netflix per country changed over time (by year) and do movies and TV shows differ in typical duration (minutes) from the following data given to us from the netflix_titles.csv file. 


1) First, we will import the data into our notebook, and then show the first 5 rows of data.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df =pd.read_csv('netflix_titles.csv')
print(df.head())
```

2) The next step will be to make sure all the data is well formatted.

```python
df["type"] = df["type"].str.strip()
df["title"] = df["title"].str.strip()
df["director"] = df["director"].str.strip()
df["cast"] = df["cast"].str.strip()
df["country"] = df["country"].str.strip()
df["date_added"] = df["date_added"].str.strip()
df["rating"] = df["rating"].str.strip()
df["duration"] = df["duration"].str.strip()
df["listed_in"] = df["listed_in"].str.strip()
df["description"] = df["description"].str.strip()

df["type"] = df["type"].str.lower()
df["title"] = df["title"].str.lower()
df["director"] = df["director"].str.lower()
df["cast"] = df["cast"].str.lower()
df["country"] = df["country"].str.lower()
df["date_added"] = df["date_added"].str.lower()
df["rating"] = df["rating"].str.lower()
df["duration"] = df["duration"].str.lower()
df["listed_in"] = df["listed_in"].str.lower()
df["description"] = df["description"].str.lower()


df["type"] = df["type"].replace({"NONE":"none"})
df["title"] = df["title"].replace({"NONE":"none"})
df["director"] = df["director"].replace({"NONE":"none"})
df["cast"] = df["cast"].replace({"NONE":"none"})
df["country"] = df["country"].replace({"NONE":"none"})
df["date_added"] = df["date_added"].replace({"NONE":"none"})
df["release_year"] = df["release_year"].replace({"NONE":"none"})
df["rating"] = df["rating"].replace({"NONE":"none"})
df["duration"] = df["duration"].replace({"NONE":"none"})
df["listed_in"] = df["listed_in"].replace({"NONE":"none"})
df["description"] = df["description"].replace({"NONE":"none"})
```

3) Now that we have fixed our data to be formatted correctly, we can create our graphs.

```python

added_per_country = df.dropna(subset=["country"]).groupby("country").size()
added_per_country.tail()
plt.figure()
plt.xticks(rotation=45)
plt.plot(added_per_country.index, added_per_country.values)
plt.xlabel("Country")
plt.ylabel("Titles added")
plt.title("Titles added to Netflix by country")
plt.show()
```

Here we can see that the US has participated the most when creating these titles that were added to Netflix.
